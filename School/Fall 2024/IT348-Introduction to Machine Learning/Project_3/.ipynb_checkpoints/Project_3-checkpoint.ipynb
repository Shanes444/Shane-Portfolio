```python
# -*- coding: utf-8 -*-
"""
Created on Fri Nov 15 15:21:46 2024

@author: Shane Stevens
"""
#code will likely be changed when collecting data to make gathering certain statistics easier

import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
import os
```

```python
folder = "input_data/"
#list of files to be used
files = ["sampled_2017_benign_api.csv","sampled_2017_malware_api.csv"]

# 0 = benign , 1 = malware
data = pd.DataFrame()
for file in files:
    temp_data = pd.read_csv(folder + file)
    if "benign" in file:
        temp_data['Target'] = 0
    elif "malware" in file:
        temp_data['Target'] = 1
    data = pd.concat([data,temp_data])
data.info()
```

```python
#create the target classes and drop columns from data
y = data['Target']
X = data.drop(['apkname','Target'], axis = 1)
X.head()
```

```python
#split data into testing and training sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
#instead of splitting into testing and training, make an entirely different testing set
folder = "input_data/"
#list of files to be used
files = ["sampled_2015_benign_api.csv","sampled_2015_malware_api.csv"]

# 0 = benign , 1 = malware
data = pd.DataFrame()
for file in files:
    temp_data = pd.read_csv(folder + file)
    if "benign" in file:
        temp_data['Target'] = 0
    elif "malware" in file:
        temp_data['Target'] = 1
    data = pd.concat([data,temp_data])
data.info()

y_train = y
X_train = X

y_test = data['Target']
X_test = data.drop(['apkname','Target'], axis = 1)
X_test.head()
```

```python
#normalize data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
print(X_train)

```

```python
#create Classification models
from sklearn.ensemble import RandomForestClassifier

#logarithmic
logarithmic = LogisticRegression(max_iter = 1000)
logarithmic.fit(X_train,y_train)

#random forest
random_forest = RandomForestClassifier(n_estimators = 500, max_depth = 4, class_weight="balanced", random_state = 42)
random_forest.fit(X_train,y_train)

#evaluate models after training
print("logarithmic:" + str(logarithmic.score(X_train,y_train)))
print("Random Forest:" + str(random_forest.score(X_train,y_train)))
```

```python
#predict testing data
from sklearn.metrics import classification_report, accuracy_score

y_pred = logarithmic.predict(X_test.values)
log_acc = accuracy_score(y_test.values, y_pred)
print("Logarithmic: " + str(log_acc))
print(classification_report(y_test.values,y_pred))

y_pred = random_forest.predict(X_test.values)
rf_acc = accuracy_score(y_test.values, y_pred)
print("Random Forest: " + str(rf_acc))
print(classification_report(y_test.values,y_pred))
if log_acc > rf_acc:
    y_pred = logarithmic.predict(X_test.values)
else:
    y_pred = logarithmic.predict(X_test.values)

```

```python
#create confusion matrix for fun
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
cm_display = ConfusionMatrixDisplay(cm, display_labels=["benign","malware"]).plot()
```
